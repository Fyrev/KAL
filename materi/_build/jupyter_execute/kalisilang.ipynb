# Kali Silang
### **Definisi Kali Silang (Cross Product)**

**Kali silang** (cross product) adalah operasi dalam aljabar vektor yang menghasilkan sebuah vektor baru yang **tegak lurus (ortogonal)** terhadap dua vektor input dalam ruang tiga dimensi. Operasi ini hanya berlaku untuk **vektor berdimensi 3**.



---

### **Rumus Kali Silang**

Misalkan terdapat dua vektor:

$$
\mathbf{u} = (u_1, u_2, u_3), \quad \mathbf{v} = (v_1, v_2, v_3)
$$

Maka hasil kali silang $\mathbf{u} \times \mathbf{v}$ adalah:

$$
\mathbf{u} \times \mathbf{v} = 
\begin{vmatrix}
\mathbf{i} & \mathbf{j} & \mathbf{k} \\
u_1 & u_2 & u_3 \\
v_1 & v_2 & v_3 \\
\end{vmatrix}
=
(u_2 v_3 - u_3 v_2)\mathbf{i}
- (u_1 v_3 - u_3 v_1)\mathbf{j}
+ (u_1 v_2 - u_2 v_1)\mathbf{k}
$$

---

### **Contoh Soal:**

Hitung $\mathbf{u} \times \mathbf{v}$ dengan:

$$
\mathbf{u} = (1, 0, 0), \quad \mathbf{v} = (0, 1, 0)
$$

Gunakan rumus:

* $u_1 = 1, u_2 = 0, u_3 = 0$
* $v_1 = 0, v_2 = 1, v_3 = 0$

Hitung tiap komponennya:

* Komponen **i**: $u_2 v_3 - u_3 v_2 = 0 \cdot 0 - 0 \cdot 1 = 0$
* Komponen **j**: $-(u_1 v_3 - u_3 v_1) = -(1 \cdot 0 - 0 \cdot 0) = 0$
* Komponen **k**: $u_1 v_2 - u_2 v_1 = 1 \cdot 1 - 0 \cdot 0 = 1$

Jadi:

$$
\mathbf{u} \times \mathbf{v} = (0, 0, 1)
$$

---

### **Hasil:**

$$
\boxed{\mathbf{u} \times \mathbf{v} = \mathbf{k} = (0, 0, 1)}
$$

---

* $\mathbf{u} = [2, -1, 4]$
* $\mathbf{v} = [3, 2, 5]$

 $\mathbf{u} \times \mathbf{v}$

Komponen-komponen:

* $u_1 = 2$, $u_2 = -1$, $u_3 = 4$
* $v_1 = 3$, $v_2 = 2$, $v_3 = 5$

$$
\mathbf{u} \times \mathbf{v} = 
\begin{bmatrix}
(-1)(5) - (4)(2) \\
(4)(3) - (2)(5) \\
(2)(2) - (-1)(3)
\end{bmatrix}
=
\begin{bmatrix}
-5 - 8 \\
12 - 10 \\
4 + 3
\end{bmatrix}
=
\begin{bmatrix}
-13 \\
2 \\
7
\end{bmatrix}
$$
